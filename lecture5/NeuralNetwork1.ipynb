```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
class NeuralNetwork:
    def __init__(self):
        self.w1_11 = None
        self.w1_12 = None
        self.b1_1 = None
        self.w1_21 = None
        self.w1_22 = None
        self.b1_2 = None
        self.w2_11 = None
        self.w2_12 = None
        self.b2_1 = None
        
    def _predict(self, x_):
        z1_1_ = self.w1_11*x_[0] + self.w1_12*x_[1] + self.b1_1
        z1_2_ = self.w1_21*x_[0] + self.w1_22*x_[1] + self.b1_2
        o1_1_ = self.tanh(z1_1_)
        o1_2_ = self.tanh(z1_2_)
        z2_1_ = self.w2_11*o1_1_ + self.w2_12*o1_2_ + self.b2_1
        o2_1_ = self.tanh(z2_1_)
        return o2_1_
    
    def tanh(self, x):
        return (np.exp(x)-np.exp(-x))/(np.exp(x) + np.exp(-x))
    
    def predict(self, x):
        return np.array([self._predict(x[i]) for i in range(len(x))])
    
    def cost(self, y, t):    
        return (1/len(t))*np.sum((y-t)**2) #Mean square error
    
    def train(self, x,t, lr=0.01, alpha =0.9):
        #convert to numpy array
        x = np.array(x)
        t = np.array(t)
        
        
        #initialization weight
        self.w1_11 = np.random.rand(1)[0]
        self.w1_12 = np.random.rand(1)[0]
        self.b1_1 = np.random.rand(1)[0]
        self.w1_21 = np.random.rand(1)[0]
        self.w1_22 = np.random.rand(1)[0]
        self.b1_2 = np.random.rand(1)[0]
        self.w2_11 = np.random.rand(1)[0]
        self.w2_12 = np.random.rand(1)[0]
        self.b2_1 = np.random.rand(1)[0]
        d_w1_11 = 0
        d_w1_12 = 0
        d_b1_1 = 0
        d_w1_21 = 0
        d_w1_22 = 0
        d_b1_2 = 0
        d_w2_11= 0 
        d_w2_12 =0 
        d_b2_1 = 0
        
        cost_list = []
        epoch_list = []
        epoch = 0
        
        while True:
            z1_1 = np.array([])
            z1_2 = np.array([])
            o1_1 = np.array([])
            o1_2 = np.array([])
            z2_1 = np.array([])
            o2_1 = np.array([])
            for i in range(len(t)):
                
                x_ = x[i]
                z1_1_ = self.w1_11*x_[0] + self.w1_12*x_[1] + self.b1_1
                z1_2_ = self.w1_21*x_[0] + self.w1_22*x_[1] + self.b1_2
                o1_1_ = self.tanh(z1_1_)
                o1_2_ = self.tanh(z1_2_)
                z2_1_ = self.w2_11*o1_1_ + self.w2_12*o1_2_ + self.b2_1
                o2_1_ = self.tanh(z2_1_)
                z1_1 =np.append(z1_1,z1_1_)
                z1_2_=np.append(z1_2,z1_2_)
                o1_1_=np.append(o1_1,o1_1_)
                o1_2_=np.append(o1_2,o1_2_)
                z2_1 = np.append(z2_1,z2_1_)
                o2_1 = np.append(o2_1,o2_1_)
                
                #gradient
                N = len(t)
                g_w1_11 = (1/N)*np.sum(2*(o2_1-t)*(1-(self.tanh(z2_1)**2))*self.w2_11*(1-(self.tanh(z1_1)**2)*x[:,0]))
                g_w1_12 = (1/N)*np.sum(2*(o2_1-t)*(1-(self.tanh(z2_1)**2))*self.w2_11*(1-(self.tanh(z1_1)**2)*x[:,1]))
                g_w1_21 = (1/N)*np.sum(2*(o2_1-t)*(1-(self.tanh(z2_1)**2))*self.w2_12*(1-(self.tanh(z1_2)**2)*x[:,0]))
                g_w1_22 = (1/N)*np.sum(2*(o2_1-t)*(1-(self.tanh(z2_1)**2))*self.w2_12*(1-(self.tanh(z1_2)**2)*x[:,1]))
                g_b1_1 = (1/N)*np.sum(2*(o2_1-t)*(1-(self.tanh(z2_1)**2))*self.w2_11*(1-(self.tanh(z1_1)**2)))
                g_b1_2 = (1/N)*np.sum(2*(o2_1-t)*(1-(self.tanh(z2_1)**2))*self.w2_12*(1-(self.tanh(z1_1)**2)))
                g_b2_1 = (1/N)*np.sum(2*(o2_1-t)*(1-(self.tanh(z2_1)**2)))
                g_w2_11 = (1/N)*np.sum(2*(o2_1-t)*(1-(self.tanh(z2_1)**2))*o1_1)
                g_w2_12 = (1/N)*np.sum(2*(o2_1-t)*(1-(self.tanh(z2_1)**2))*o1_2)
                
                #compute cost
                cost = self.cost(o2_1,t)
                cost_list.append(cost)
                epoch_list.append(epoch)
                epoch +=1
                print(f"epoch = {epoch}, cost = {cost}")
                
                
                
                #compute momentum
                d_w1_11 = lr*g_w1_11 + alpha*d_w1_11
                d_w1_12 = lr*g_w1_12 + alpha*d_w1_12
                d_b1_1 = lr*g_b1_1 + alpha*d_b1_1                
                d_w1_21 = lr*g_w1_21 + alpha*d_w1_21
                d_w1_22 = lr*g_w1_22 + alpha*d_w1_22
                d_b1_2 = lr*g_b1_2 + alpha*d_b1_2
                d_w2_11= lr*g_w2_11 + alpha*d_w2_11
                d_w2_12 =lr*g_w2_12 + alpha*d_w2_12
                d_b2_1 = lr*g_b2_1 + alpha*d_b2_1

                #update weight
                self.w1_11 = self.w1_11 - d_w1_11
                self.w1_12 = self.w1_12 - d_w1_12
                self.b1_1 = self.b1_1 - d_b1_1
                self.w1_21 = self.w1_21 - d_w1_21
                self.w1_22 = self.w1_22 - d_w1_22
                self.b1_2 = self.b1_2 - d_b1_2
                self.w2_11 = self.w2_11 - d_w2_11
                self.w2_12 = self.w2_12 - d_w2_12
                self.b2_1 = self.b2_1 - d_b2_1
    if __name__ == '__main__':

        #set parameter
        data_filename = "data.mat"
        lr = 0.01
        alpha = 0.9

        #load data
        data = sio.loadmat(data_filename)
        x = data['x']
        t = data['t'][0]

        #create object
        nn = NeuralNetwork()
        nn.train(x,t)
                                                                                    
            
                
        
```
